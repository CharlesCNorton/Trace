```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
checks = [1,2,3,4,5,6]
feedbacks = [0,1]
contexts = [0,1]
interactions = [
    [
        [2,5,9,53,89,117],
        [2,3,3,8,9,107]
    ],
    [
        [1,7,2,12,56,81],
        [0,0,0,0,0,0]
    ]
]
```

```python
# Plot using seaborn
sns.set_style("whitegrid")
palette = sns.color_palette("bright")
sns.set_palette(palette)
plt.figure()
sns.lineplot(x=checks, y=interactions[0][0], label='original feedback', marker='s')
sns.lineplot(x=checks, y=interactions[0][1], label='structured feedback', marker='o')
plt.ylabel('interactions')
plt.xlabel('checks')
plt.title('#interactions vs. #checks with default context')
plt.legend(loc='upper left')
plt.show()
```

```python
# Plot using seaborn
sns.set_style("whitegrid")
palette = sns.color_palette("bright")
sns.set_palette(palette)
plt.figure()
sns.lineplot(x=checks, y=interactions[0][0], label='default context', marker='s')
sns.lineplot(x=checks, y=interactions[1][0], label='structured context', marker='o')
plt.ylabel('interactions')
plt.xlabel('checks')
plt.title('#interactions vs. #checks with original feedback')
plt.legend(loc='upper left')
plt.show()
```
